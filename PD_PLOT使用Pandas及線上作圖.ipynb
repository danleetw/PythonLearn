### 從Nasdaq讀資料

```python
import pandas as pd

url="https://www.nasdaq.com/screening/companies-by-industry.aspx?exchange=NASDAQ&render=download"

company = pd.read_csv(url)
company.head()
```

### 將資料轉成List

```python
import pandas_datareader as web
from datetime import datetime
symbols=company["Symbol"][0:20].tolist()

symbols
```

### 讀取前述公司股票資料

```python
data={}
symbols={'ABMD'}
symbols={'2317.TW'}
for symbol in symbols:
    try:
        data[symbol]=web.DataReader(symbol,"yahoo",datetime(2012,1,1))
        
    except Exception as e:

        print("Error occurs while access symbol:",symbol," --",e)
```

### 測試抓到的資料

```python
data['2317.TW']
```

```python
%matplotlib inline
data.keys()

data['2317.TW'][['Close']].plot()
data['2317.TW'][['Open']].plot()
```
